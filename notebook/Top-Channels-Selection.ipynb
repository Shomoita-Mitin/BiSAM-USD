# 1. Loading the EEG Signal Data

```python
import os
import mne
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_folder = 'D:\Thesis Dataset\IOWA Data\All_Rest_Recordings_RAW\All_Rest_Recordings_RAW'

all_files = os.listdir(data_folder)

vhdr_files = [file for file in all_files if file.endswith('.vhdr')]

raw_list = []

for vhdr_file in vhdr_files:
    file_path = os.path.join(data_folder, vhdr_file)
    raw = mne.io.read_raw_brainvision(file_path, preload=True)
    raw_list.append(raw)
```

# 3. Model Training based on the anatomy of the EEG Signal: Epochs vs Channels

```python
sampling_rate = int(raw.info['sfreq']) 
epoch_length = 1 
n_samples_per_epoch = sampling_rate * epoch_length

n_epochs = eeg_data.shape[1] // n_samples_per_epoch

eeg_data_epochs = eeg_data[:, :n_epochs * n_samples_per_epoch].reshape(len(eeg_channel_names), n_epochs, n_samples_per_epoch)
eeg_data_epochs = np.swapaxes(eeg_data_epochs, 0, 1)  

X_channels_feature = np.mean(eeg_data_epochs, axis=2) 

n_epochs = X_channels_feature.shape[0] 

labels = ['eyesopen'] * 90 + ['eyesclosed'] * 93 
label_mapping = {'eyesopen': 0, 'eyesclosed': 1}
y_channels_feature = np.array([label_mapping[label] for label in labels])

print(f"Shape of X_channels: {X_channels_feature.shape}")
print(f"Shape of y_chennels: {y_channels_feature.shape}")
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

X_chn_train, X_chn_test, y_chn_train, y_chn_test = train_test_split(X_channels_feature, y_channels_feature, test_size=0.2, random_state=42)

xgb_model_chn = XGBClassifier()
xgb_model_chn.fit(X_chn_train, y_chn_train)

importances_chn = xgb_model_chn.feature_importances_

eeg_channel_names = [raw.info['ch_names'][i] for i in mne.pick_types(raw.info, eeg=True)]
importance_df = pd.DataFrame({
    'Channel': eeg_channel_names,
    'Importance': importances_chn
})

importance_df = importance_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)
print("Channel Importances:\n", importance_df)
```

```python
plt.figure(figsize=(12, 8))
plt.barh(importance_df['Channel'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Channel')
plt.title('EEG Channel Importance')
plt.gca().invert_yaxis()  
plt.show()
```

# 4. Channels Selection: Selecting the top channels with the importance value

```python
top_16_channels = importance_df.head(16)

print("Top 16 Channels by Importance:\n", top_16_channels)
```

```python
top_8_channels = importance_df.head(8)

print("Top 8 Channels by Importance:\n", top_8_channels)
```

```python
top_4_channels = importance_df.head(4)

print("Top 4 Channels by Importance:\n", top_4_channels)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mne

top_channels = {
    'Top 16 Channels': importance_df.head(16),
    'Top 8 Channels': importance_df.head(8),
    'Top 4 Channels': importance_df.head(4)
}

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 15), sharex=True)
fig.suptitle('Importance of the Top Channels', fontsize=16)

for ax, (title, df) in zip(axes, top_channels.items()):
    channels = df['Channel'].tolist()
    indices = [eeg_channel_names.index(ch) for ch in channels]
    
    data = raw.get_data(picks=indices)
    
    for i, channel_index in enumerate(indices):
        ax.plot(raw.times, data[i, :], label=channels[i])
    
    ax.set_title(title)
    ax.set_ylabel('Importance')
    ax.legend(loc='upper right')

axes[-1].set_xlabel('Time (s)')
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
